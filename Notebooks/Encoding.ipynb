```python
def technology(t):
    if t=="Frontend using HTML, CSS and JavaScript.":
        return 1
    elif t=="Frontend using React.":
        return 2
    elif t=="Backend using JavaScript frameworks.":
        return 3
    elif t=="Backend using Python Framework.":
        return 4
    elif t=="Backend using Java Frameworks.":
        return 5
    elif t=="Competitive Programming using C++":
        return 6
    elif t=="Competitive Programming using Python":
        return 7
    elif t=="Competitive Programming using Java":
        return 8
    elif t=="Full Stack (JavaScript Based Tech Stack).":
        return 9
    elif t=="Full Stack (Python Based Tech Stack).":
        return 10
    elif t=="Full Stack (Java Based Tech Stack).":
        return 11
    elif t=="Data Science / AI / ML":
        return 12
    elif t=="Blockchain":
        return 13
    elif t=="Cyber Security":
        return 14
    elif t=="Application Development using Kotlin":
        return 15
    elif t=="Application Development using Java":
        return 16
    elif t=="Application Development using Swift":
        return 17
    elif t=="Application Development using Flutter":
        return 18
    else:
        return 19
```

```python
df["curr_techno"]=df.curr_techno.apply(technology)
```

```python
def Year(y):
    if y=="Second":
        return 2
    elif y=="Third":
        return 3
    elif y=="Fourth":
        return 4
```
